```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from torchvision import datasets, transforms

# Define dataset path
data_dir = "../data/Animals-10/raw-img"

# Check dataset folders
categories = os.listdir(data_dir)
print(f"Categories: {categories}")

# Count images per category
category_counts = {category: len(os.listdir(os.path.join(data_dir, category))) for category in categories}

# Visualize category distribution
plt.figure(figsize=(10, 5))
sns.barplot(x=list(category_counts.keys()), y=list(category_counts.values()))
plt.xticks(rotation=45)
plt.xlabel("Animal Category")
plt.ylabel("Number of Images")
plt.title("Dataset Distribution")
plt.show()

# Show sample images
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
valid_categories = [c for c in categories if len(os.listdir(os.path.join(data_dir, c))) > 0]  # Filter out empty folders
for ax, category in zip(axes.flatten(), valid_categories):
    img_list = os.listdir(os.path.join(data_dir, category))
    if img_list:  # Ensure the folder is not empty
        img_path = os.path.join(data_dir, category, img_list[0])
        img = Image.open(img_path)
        ax.imshow(img)
        ax.set_title(category)
        ax.axis("off")
    else:
        ax.axis("off")  # Hide axes if no image available
plt.tight_layout()
plt.show()
```
